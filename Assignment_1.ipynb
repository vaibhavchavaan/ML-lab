```python
#Step 3: Data Reduction

# Remove S.No. column from data
data = data.drop(['S.No.'], axis = 1)
data.info()
```

```python
#Step 4&5: Creating Features

from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
data['Brand'] = data.Name.str.split().str.get(0)
```

```python
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
```

```python
data[['Name','Brand','Model']]
```

```python
from datetime import datetime

# Current year
current_year = datetime.now().year

# Create a new column 'car_age'
data['car_age'] = current_year - data['Year']

# Display updated DataFrame
data[['Year', 'car_age']].head()
```

```python
# display first 10 records
data.head(10)
```

```python
#Step 6:Data Cleaning and Wrangling

print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
data["Brand"].replace({"Land":"Range Rover"}, inplace=True)
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
searchfor = ['Fiat','Land ']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
data["Brand"].replace({"Fiat": "Fiat Panda","Land":"Range Rover"}, inplace=True)
```

```python
import seaborn as sns
import pandas as pd

# Example DataFrame for demonstration
data = pd.DataFrame({
	'fuel_type': ['Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel'],
	'price': [5.2, 3.8, 7.1, 4.5, 6.0, 5.5, 8.2, 4.0, 6.5, 7.8]
})

sns.barplot(x='fuel_type', y='price', data=data)
plt.title('Average Price by Fuel Type')
plt.xticks(rotation=45)
plt.show()
```

```python
data.hist(figsize=(12, 8), edgecolor='black')
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.title("Boxplot of All Numeric Features")
plt.show()
```

```python
# Create a DataFrame for brands and avg_prices
brand_df = pd.DataFrame({'brand': brands, 'price': avg_prices})

sns.barplot(x='brand', y='price', data=brand_df)
plt.title("Average Price by Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
# 'data' only contains 'price' and 'fuel_type', so we can only plot 'price'
sns.pairplot(data[['price']])
plt.suptitle("Pair Plot of Price Feature", y=1.02)
plt.show()
```
